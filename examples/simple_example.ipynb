This is a simple example on how to use pyDVC given a Numpy array of positions and weights

```python
import pyDVC
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

Establish how many particles we want to tessellate

```python
n_particles = 1000
```

Randomly generate the positions and weights of the particles

```python
positions = np.random.rand(n_particles, 3) - 0.5
weights = np.random.rand(n_particles) * 0.05
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')

ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2])
ax.set_title('Positions of particles')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

Decide the rectangular bounds of the tessellation (same unit as positions)

```python
bounds = np.array([
    [-0.5, 0.5],
    [-0.5, 0.5],
    [-0.5, 0.5]
])
```

Perform the actual tessellation

```python
ownership_array = pyDVC.tessellate(positions=positions,
                                   bounds=bounds,
                                   cell_length=1/1000,  # 1000 cells in each axis
                                   radii=weights)
```

Plot the result

```python
fig, ax = plt.subplots()
ax.imshow(ownership_array[50, :, :])
plt.show()
```
